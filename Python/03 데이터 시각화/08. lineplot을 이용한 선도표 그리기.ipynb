# **lineplot 을 이용한 선도표 그리기**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

```python
df.head()
```

\- 결측값 확인

```python
df.isnull().sum()
```

\- 선도표 그리기

```python
sns.lineplot(data = df, x = 'year', y = 'passengers')
```

\- xtick 에 전체 년도 표시

```python
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year']);       # ;를 생략하면 텍스트들도 다 출력됨 시각적으로 좋게하기 위해 ; 추가
```

\- xtick 기울기 추가

```python
sns.lineplot(data = df, x = 'year', y = 'passengers')
plt.xticks(df['year'], rotation = 45);
```

\- 월별로 그리기

```python
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month')
```

\- 색상 변경([옵션 링크](https://seaborn.pydata.org/tutorial/color_palettes.html))

```python
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2')
```

\- 선 표현 방법 변경

```python
sns.lineplot(data = df, x = 'year', y = 'passengers', hue = 'month', palette = 'Set2', style = 'month')
```
