```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

```python
df = sns.load_dataset('flights')
```

데이터 형태 확인

```python
df.head()
```

```python
df.describe()
```

결측값 확인

\- 문제 : 각 년도별 월별 승객수를 히트맵으로 나타내시오.

pivot 을 사용한 데이터 형태 변경

```python
pivot_data = df.pivot("month", "year", "passengers")
```

```python
pivot_data
```

```python
sns.heatmap(pivot_data)
```

\- 구분선 추가

```python
sns.heatmap(pivot_data, linewidths=.5)
```

\- cmap 을 통한 colorbar 색상 변경

```python
sns.heatmap(pivot_data, cmap="Blues")
```

\- 수치 입력

```python
sns.heatmap(pivot_data, cmap="Blues", annot = True)
```

\- 정수 형태로 출력

```python
sns.heatmap(pivot_data, cmap="Blues", annot = True, fmt = 'd')
```

\- colorbar 가로로 놓기

```python
fig, (ax, cbar_ax) = plt.subplots(2)
```

```python
fig
```

```python
ax
```

```python
cbar_ax
```

```python
ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "horizontal"})
```

```python
fig
```

orientation : vertical

```python
ax = sns.heatmap(pivot_data, ax=ax,
                  cbar_ax=cbar_ax,
                  cbar_kws={"orientation": "vertical"})
```

```python
fig
```

\- colorbar 와 히트맵 같의 비율 조정

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"})
```

\- 최종

```python
grid_kws = {"height_ratios": (.85, .1), "hspace": 0.4}

f, (ax, cbar_ax) = plt.subplots(2, gridspec_kw=grid_kws)
ax = sns.heatmap(pivot_data, ax=ax,
                 cbar_ax=cbar_ax,
                 cbar_kws={"orientation": "horizontal"},
                 cmap="Blues", 
                 annot = True, fmt = 'd')
```
