# **boxplot 을 이용한 상자 수염 그림 그리기**

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 5)})
import pandas as pd
```

\- iris 데이터 셋 로드

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

\- boxplot 그리기

```python
sns.boxplot(data = df)
```

\- 세로방향으로 그림 크기 키우기

```python
plt.subplots(figsize = (7, 8))
sns.boxplot(data = df)
```

\- 가로 방향으로 그리기

```python
sns.boxplot(data = df, orient = 'h')
```

\- swarmplot 을 이용한 raw data 확인

```python
plt.subplots(figsize = (10, 10))
sns.swarmplot(data = df)
```

\- 겹쳐 그리기

```python
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df)
```

\- swarmplot 색상 조정

```python
plt.subplots(figsize = (10, 10))
sns.boxplot(data = df)
sns.swarmplot(data = df, color = 'black')
```

\- violinplot

```python
sns.violinplot(data = df)
```

\- stripplot

```python
sns.stripplot(data = df)
```
