# **matplotlib을 이용한 단일 차트 그리기**

* library 호출

```python
import matplotlib.pyplot as plt
```

 \- 문제 : x 가 1, 2, 3 이고 y 가 4, 5, 6 인 그래프를 그리시오.

```python
# 리스트 형태로 저장

x = [1, 2, 3]
y = [4, 5, 6]
```

```python
plt.figure()        # 그림을 그려줄 도화지, 생략 가능
plt.plot(x, y)      # x, y 데이터를 입력
```

또는

```python
plt.plot(x, y)
```

\- lineplot 은 y 값만 있으면 그림을 그릴 수 있음

\- 이 때, x 의 값은 0, 1, 2... 순서로 자동 지정됨

```python
plt.plot(y)
```

* linewidth 를 이용한 선 굵기 조정

```python
plt.plot(x, y, linewidth = 10)
```

* color 를 통한 선 색상 조정

```python
plt.plot(x, y, color = 'red')       # red 대신 r로 해도 됨
```

* marker 를 이용한 데이터 위치 표시([marker 문서](https://matplotlib.org/stable/api/markers_api.html))

```python
plt.plot(x, y, marker = 'o')
```

* marker 크기 변경

```python
plt.plot(x, y, marker = 'o', markersize = 10)
```

* 선 형태 변경

```python
plt.plot(x, y, linestyle = ':')
```

```python
plt.plot(x, y, 'o')
```

```python
plt.plot(x, y, 'ro')      # 약어 ro : 빨갛고 동그랗게
```

* 그래프 명

```python
plt.plot(x, y)
plt.title('title')
```

* 그래프 명 폰트 크기 변경

```python
plt.plot(x, y)
plt.title('title', fontsize = 20)
```

* x 축 명, y 축 명

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```

* 축 명 폰트 크기 변경

```python
plt.plot(x, y)
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
```

* 그리드 설정(전체)

```python
plt.plot(x, y)
plt.grid(True)
```

* 그리드 설정(x 축만)

```python
plt.plot(x, y)
plt.grid(True, axis = 'x')
```

* 그리드 설정(y 축만)

```python
plt.plot(x, y)
plt.grid(True, axis = 'y')
```

* x 범위, y 범위 설정

```python
plt.plot(x, y)
plt.xlim([1, 2])
plt.ylim([4, 5])
```

* axis 를 사용한 범위 설정

```python
plt.plot(x, y)
plt.axis([1, 2, 4, 5])      # plt.axis([x최솟값, x최댓값, y최솟값, y최댓값])
```

* 눈금 글꼴 크기 변경

```python
plt.plot(x, y)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 20)
```

* 그래프에 텍스트 삽입

```python
plt.plot(x, y)
plt.text(2, 5, 'text', fontsize = '20')
```

* 한글폰트 사용하기

```python
plt.plot(x)
plt.title('차트 명')
```
